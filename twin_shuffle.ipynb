```python
import numpy as np
from numpy.random import rand

import pandas as pd

from sklearn.utils import shuffle

import time
import resource 
```

```python
original = rand(1000)
index = np.arange(1000)
```

```python
print('original: ', original)
print('index: ', index)
```

```python
time_start = time.perf_counter()

A, B = shuffle(original, index)

time_elapsed = (time.perf_counter() - time_start)
memMb=resource.getrusage(resource.RUSAGE_SELF).ru_maxrss/1024.0/1024.0
print ("%5.12f secs %5.1f MByte" % (time_elapsed,memMb))
```

```python
def twin_shuffle(a, b):
    rng_state = np.random.get_state()
    A = np.random.shuffle(a)
    B = np.random.set_state(rng_state)
    np.random.shuffle(b)
    return A, B
```

```python
print('original: ', original)
print('index: ', index)
```

```python
time_start = time.perf_counter()

x, y = twin_shuffle(original, index)

time_elapsed = (time.perf_counter() - time_start)
memMb=resource.getrusage(resource.RUSAGE_SELF).ru_maxrss/1024.0/1024.0
print ("%5.12f secs %5.1f MByte" % (time_elapsed,memMb))
```
